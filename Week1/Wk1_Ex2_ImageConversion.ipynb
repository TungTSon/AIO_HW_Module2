```python
!gdown 1i9dqan21DjQoG5Q_VEvm0LrVwAlXD0vB
import matplotlib.image as mpimg
import matplotlib.pyplot as plt # Import the pyplot module

img = mpimg.imread("./dog.jpeg")
plt.imshow(img) # Use plt.imshow to display the image
plt.show() # Display the plot
```

**Greyscale Conversion (Lightness method)**

```python
import numpy as np
gray_img_01 = np.max(img, axis=-1, keepdims=1)/2+np.min(img, axis=-1, keepdims=1)/2
print("Result: ", gray_img_01[0, 0])
plt.imshow(gray_img_01)
plt.show()
```

**Greyscale Conversion (Average method)**

```python
gray_img_02 = np.mean(img, axis=-1, keepdims=1)
print("Result: ", gray_img_02[0, 0])
plt.imshow(gray_img_02)
plt.show()
```

**Greyscale Conversion (Luminosity method)**

```python
# split the red, green blue channel of img into 3 variables
red_channel, green_channel, blue_channel = img[:, :, 0], img[:, :, 1], img[:, :, 2]
```

```python
gray_img_03 = 0.21 * red_channel + 0.72 * green_channel + 0.07 * blue_channel
print("Result: ", gray_img_03[0, 0])
plt.imshow(gray_img_03)
plt.show()
```
